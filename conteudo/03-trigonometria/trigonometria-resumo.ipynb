## Funções Trigonométricas

### Função Seno

```python
import numpy as np
import math

x_values = []
y_values = []
for x in np.arange(-4*math.pi, 4*math.pi, 0.1):
    x_values.append(x)
    y_values.append(math.sin(x))

plt.axhline(y=0.0, color='grey', linestyle='-', linewidth=1)
plt.axvline(x=0.0, color='grey', linestyle='-', linewidth=1)
plt.plot(x_values, y_values)
plt.show()
```

### Função Cosseno

```python
import numpy as np
import math

x_values = []
y_values = []
for x in np.arange(-4*math.pi, 4*math.pi, 0.1):
    x_values.append(x)
    y_values.append(math.cos(x))

plt.axhline(y=0.0, color='grey', linestyle='-', linewidth=1)
plt.axvline(x=0.0, color='grey', linestyle='-', linewidth=1)
plt.ylim(-1, 1)
plt.plot(x_values, y_values)
plt.show()
```

### Função Tangente

```python
import numpy as np
import math

x_min = -2 * math.pi
x_max = 2 * math.pi
x_step = (x_max - x_min) / 1000
x_values = np.arange(x_min, x_max + x_step, x_step)
y_values = np.tan(x_values)
y_values[:-1][np.diff(y_values) < 0] = np.nan

plt.grid()
plt.axhline(y=0.0, color='grey', linestyle='-', linewidth=1)
plt.axvline(x=0.0, color='grey', linestyle='-', linewidth=1)
plt.xlim(x_min, x_max)
plt.xlabel("$x$ [rad]", fontsize=14)
plt.ylabel("$tg(x)$", fontsize=14)
plt.ylim(-10, 10)
radian_multiples = np.arange(x_min / math.pi, x_max / math.pi + 0.5, 0.5) # [-2, -3/2, -1, -1/2, 0, 1/2, 1, 3/2]
radians = [n * math.pi for n in radian_multiples]
radian_labels = []
for n in radian_multiples:
    label = '0' if n == 0 else rf'{int(n)}$\pi$' if n == int(n) else rf'{n}$\pi$'
    radian_labels.append(label)

plt.xticks(radians, radian_labels)
plt.plot(x_values, y_values)
plt.show()
```
