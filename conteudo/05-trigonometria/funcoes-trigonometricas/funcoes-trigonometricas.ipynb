# 2. <a id='toc2_'></a>[Funções Trigonométricas](#toc0_)

## 2.1. <a id='toc2_1_'></a>[Definição](#toc0_)

__Funções trigonométricas__ são aplicações de $\mathbb{R}$ sobre o ciclo trigonométrico $\lambda$, onde podemos associar a cada número real $x$ um único ponto $P$ de $\lambda$ do seguinte modo:

1. Se $\,x = 0\,$, então $P$ coincide com $A$;

2. Se $\,x > 0\,$, então realizamos a partir de $A$ um percurso de comprimento $x$, no sentido anti-horário, e marcamos $P$ como ponto final do percurso;

3. Se $\,x < 0\,$, então realizamos a partir de $A$ um percurso de comprimento $|x|$, no
sentido horário. O ponto final do percurso é $P$.

$$
\{x \in \mathbb{R}\ | \ x = x_0 + 2k\pi, \ k \in \mathbb{Z}\}

# 3. <a id='toc3_'></a>[Função Seno: $\ y = \text{sen }x\ $](#toc0_)

## 3.1. <a id='toc3_1_'></a>[Definição](#toc0_)

Denominamos __função seno__ a função $\,f: \mathbb{R} \longrightarrow \mathbb{R}\,$ que associa a cada
real $x$ o real $\,OP_1 = \text{sen }x\,$, isto é:

$$
f(x) = \text{sen }x
$$

## 3.2. <a id='toc3_2_'></a>[Gráfico](#toc0_)

```python
import numpy as np
import math

x_min = -2 * math.pi
x_max = 2 * math.pi
x_step = (x_max - x_min) / 1000
y_min = -1
y_max = 1
plt.xlabel("$x$ [rad]", fontsize=14)
plt.ylabel("$sin(x)$", fontsize=14)
x_values = np.arange(x_min, x_max + x_step, x_step)
y_values = np.sin(x_values)

plt.grid()
plt.axhline(y=0.0, color='grey', linestyle='-', linewidth=1)
plt.axvline(x=0.0, color='grey', linestyle='-', linewidth=1)
plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
radian_multiples = np.arange(x_min / math.pi, x_max / math.pi + 0.5, 0.5) # [-2, -3/2, -1, -1/2, 0, 1/2, 1, 3/2]
radians = [n * math.pi for n in radian_multiples]
radian_labels = []
for n in radian_multiples:
    label = '0' if n == 0 else rf'{int(n)}$\pi$' if n == int(n) else rf'{n}$\pi$'
    radian_labels.append(label)

plt.xticks(radians, radian_labels)
plt.plot(x_values, y_values)
plt.show()
```

# 4. <a id='toc4_'></a>[Função Cosseno: $\ y = \cos x\ $](#toc0_)

## 4.1. <a id='toc4_1_'></a>[Definição](#toc0_)


Denominamos __função cosseno__ a função $\,f: \mathbb{R} \longrightarrow \mathbb{R}\,$ que associa a cada
real $x$ o real $\,OP_2 = \cos x\,$, isto é:

$$
f(x) = \cos x
$$

## 4.2. <a id='toc4_2_'></a>[Gráfico](#toc0_)

```python
import numpy as np
import math

x_min = -2 * math.pi
x_max = 2 * math.pi
x_step = (x_max - x_min) / 1000
x_values = np.arange(x_min, x_max + x_step, x_step)
y_values = np.cos(x_values)

plt.grid()
plt.axhline(y=0.0, color='grey', linestyle='-', linewidth=1)
plt.axvline(x=0.0, color='grey', linestyle='-', linewidth=1)
plt.xlim(x_min, x_max)
plt.xlabel("$x$ [rad]", fontsize=14)
plt.ylabel("$cos(x)$", fontsize=14)
plt.ylim(-1, 1)
radian_multiples = np.arange(x_min / math.pi, x_max / math.pi + 0.5, 0.5) # [-2, -3/2, -1, -1/2, 0, 1/2, 1, 3/2]
radians = [n * math.pi for n in radian_multiples]
radian_labels = []
for n in radian_multiples:
    label = '0' if n == 0 else rf'{int(n)}$\pi$' if n == int(n) else rf'{n}$\pi$'
    radian_labels.append(label)

plt.xticks(radians, radian_labels)
plt.plot(x_values, y_values)
plt.show()
```

# 5. <a id='toc5_'></a>[Função Tangente: $\ y = \text{tg }x\ $](#toc0_)

Denominamos __função tangente__ a função $\,f: D \longrightarrow \mathbb{R}\,$, que associa a cada
real $x$, $\,x \neq \cfrac{\pi}{2} + k\pi\,$, o real $\,AT = \text{tg }x\,$, isto é:

$$
f(x) = \text{tg }x
$$

## 5.1. <a id='toc5_1_'></a>[Propriedades](#toc0_)

1. O domínio da função tangente é $\,D = \left\{ x \in \mathbb{R} \ | \ x \neq \cfrac{\pi}{2} + k\pi, \ k \in \mathbb{Z} \right\}\,$.

2. A imagem da função tangente é $\mathbb{R}$, isto é, para todo $y$ real existe um $x$ real tal que
$\,\text{tg }x = y\,$.
De fa

3. A função tangente é periódica e seu período é $\pi$.

$$
\text{tg }x = \text{tg }(x + k\pi) 
$$

## 5.2. <a id='toc5_2_'></a>[Gráfico](#toc0_)

```python
import numpy as np
import math

x_min = -2 * math.pi
x_max = 2 * math.pi
x_step = (x_max - x_min) / 1000
x_values = np.arange(x_min, x_max + x_step, x_step)
y_values = np.tan(x_values)
y_values[:-1][np.diff(y_values) < 0] = np.nan

plt.grid()
plt.axhline(y=0.0, color='grey', linestyle='-', linewidth=1)
plt.axvline(x=0.0, color='grey', linestyle='-', linewidth=1)
plt.xlim(x_min, x_max)
plt.xlabel("$x$ [rad]", fontsize=14)
plt.ylabel("$tg(x)$", fontsize=14)
plt.ylim(-10, 10)
radian_multiples = np.arange(x_min / math.pi, x_max / math.pi + 0.5, 0.5) # [-2, -3/2, -1, -1/2, 0, 1/2, 1, 3/2]
radians = [n * math.pi for n in radian_multiples]
radian_labels = []
for n in radian_multiples:
    label = '0' if n == 0 else rf'{int(n)}$\pi$' if n == int(n) else rf'{n}$\pi$'
    radian_labels.append(label)

plt.xticks(radians, radian_labels)
plt.plot(x_values, y_values)
plt.show()
```
