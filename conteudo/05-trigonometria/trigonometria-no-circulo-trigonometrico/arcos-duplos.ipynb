# 1. <a id='toc1_'></a>[Soma e Subtração de Arcos](#toc0_)

## 1.1. <a id='toc1_1_'></a>[Seno](#toc0_)

$$
\text{sen}\,(a + b) = \text{sen}\,a\cdot\cos{b} + \text{sen}\,b\cdot\cos{a}
$$

&emsp;&emsp; A função cosseno é par, ou seja, $\,\cos{(-x)} = \cos{x}\,$. Já a função seno é par, ou seja, $\text{sen}\,(-x) = -\text{sen}\,x$.

&emsp;&emsp; Então,

$$
\text{sen}\,(a - b) = \text{sen}\,[a + (-b)] = \text{sen}\,a\cdot\cos{(-b)} + \text{sen}\,(-b)\cdot\cos{a}
$$

$$
\therefore\quad \text{sen}\,(a - b) = \text{sen}\,a\cdot\cos{b} - \text{sen}\,b\cdot\cos{a}
$$

&emsp;&emsp; Podemos escrever essa duas identidades de forma única:

$$
\boxed{\,\text{sen}\,(a \pm b) = \text{sen}\,a\cdot\cos{b} \pm \text{sen}\,b\cdot\cos{a}\,}
$$

&emsp;&emsp; Exemplo:

&emsp;&emsp; Calcular o valor do $\text{sen}\,15°$.

$$
\text{sen}\,15° = \text{sen}\,(45° - 30°) = \text{sen}\,45°\cdot\cos{30°} - \text{sen}\,30°\cdot\cos{45°}
$$

$$
\text{sen}\,15° = \cfrac{\sqrt{2}}{2}\cdot\cfrac{\sqrt{3}}{2} - \cfrac{1}{2}\cdot\cfrac{\sqrt{2}}{2}
$$

$$
\text{sen}\,15° = \cfrac{\sqrt{2}\cdot\sqrt{3}}{4} - \cfrac{\sqrt{2}}{4}
$$

$$
\text{sen}\,15° = \cfrac{\sqrt{6}-\sqrt{2}}{4}
$$

## 1.2. <a id='toc1_2_'></a>[Cosseno](#toc0_)

$$
\cos{(a + b)} = \cos{a}\cdot\cos{b} - \text{sen}\,a\cdot\text{sen}\,b
$$

&emsp;&emsp; Para determinarmos $\cos{(a - b)}$, devemos lembrar que a função cosseno é par, ou seja, $\,\cos{(-x)} = \cos{x}\,$ e a função seno é ímpar, ou seja, $\text{sen}\,(-x) = -\text{sen}\,x$.

&emsp;&emsp; Então,

$$
\cos{(a - b)} \,=\, \cos{[a + (-b)]} = \cos{a}\cdot\cos{(-b)} - \text{sen}\,(a)\cdot\text{sen}\,(-b)
$$

$$
\therefore\quad \cos\,(a - b) = \cos{a}\cdot\cos{b} + \text{sen}\,a\cdot\text{sen}\,b
$$

&emsp;&emsp; Podemos escrever essa duas identidades de forma única:

$$
\boxed{\,\cos{(a \pm b)} = \cos{a}\cdot\cos{b} \mp \text{sen}\,a\cdot\text{sen}\,b\,}
$$

&emsp;&emsp; Exemplo:

&emsp;&emsp; Calcular o valor do $\cos{15°}$.

$$
\cos{15°} = \cos{(45° - 30°)} = \cos{45°}\cdot\cos{30°} + \text{sen}\,45°\cdot\text{sen}\,{30°}
$$

$$
\cos{15°} = \cfrac{\sqrt{2}}{2}\cdot\cfrac{\sqrt{3}}{2} + \cfrac{\sqrt{2}}{2}\cdot\cfrac{1}{2}
$$

$$
\cos{15°} = \cfrac{\sqrt{6}}{4} + \cfrac{\sqrt{2}}{4}
$$

$$
\cos{15°} = \cfrac{\sqrt{6}+\sqrt{2}}{4}
$$

## 1.3. <a id='toc1_3_'></a>[Tangente](#toc0_)

$$
\text{tg}\,{(a + b)} = \cfrac{\text{sen}\,(a + b)}{\cos{(a + b)}}
$$

$$
\text{tg}\,{(a + b)} = \cfrac{\text{sen}\,a\cdot\cos{b} + \text{sen}\,b\cdot\cos{a}}{\cos{a}\cdot\cos{b} - \text{sen}\,a\cdot\text{sen}\,b}
$$

&emsp;&emsp; Dividindo tudo por $\cos{a}\cdot\cos{b}$, temos

$$
\text{tg}\,{(a + b)} = \cfrac{\ \cfrac{\text{sen}\,a\cdot\cos{b} + \text{sen}\,b\cdot\cos{a}}{\cos{a}\cdot\cos{b}}\ }{\ \cfrac{\cos{a}\cdot\cos{b} - \text{sen}\,a\cdot\text{sen}\,b}{\cos{a}\cdot\cos{b}}\ }
$$

$$
\text{tg}\,{(a + b)} = \cfrac{\ \cfrac{\text{sen}\,a\cdot\cos{b}}{\cos{a}\cdot\cos{b}} + \cfrac{\text{sen}\,b\cdot\cos{a}}{\cos{a}\cdot\cos{b}}\ }{\ \cfrac{\cos{a}\cdot\cos{b}}{\cos{a}\cdot\cos{b}} - \cfrac{\text{sen}\,a\cdot\text{sen}\,b}{\cos{a}\cdot\cos{b}}\ }
$$


&emsp;&emsp; Então,

$$
\text{tg}\,{(a + b)} = \cfrac{\ \text{tg}\,a + \text{tg}\,{b}}{\ 1 - \text{tg}\,a\cdot\text{tg}\,b}
$$

&emsp;&emsp; Analogamente,

$$
\text{tg}\,{(a - b)} = \cfrac{\ \text{tg}\,a - \text{tg}\,{b}}{\ 1 + \text{tg}\,a\cdot\text{tg}\,b}
$$

Portanto,

$$
\boxed{\,\text{tg}\,{(a \pm b)} = \cfrac{\ \text{tg}\,a \pm \text{tg}\,{b}}{\ 1 \mp \text{tg}\,a\cdot\text{tg}\,b}\,}
$$

# 2. <a id='toc2_'></a>[Arcos Duplos](#toc0_)

## 2.1. <a id='toc2_1_'></a>[Seno](#toc0_)

&emsp;&emsp; Vimos que

$$
\text{sen}\,(a + b) = \text{sen}\,a\cdot\cos{b} + \text{sen}\,b\cdot\cos{a}
$$

&emsp;&emsp; Fazendo $\,a = x\,$ e $\,b = x\,$, temos

$$
\text{sen}\,(x + x) = \text{sen}\,x\cdot\cos{x} + \text{sen}\,x\cdot\cos{x}
$$

$$
\boxed{\,\text{sen}\,(2x) = 2\,\text{sen}\,x\cdot\cos{x}\,}
$$



## 2.2. <a id='toc2_2_'></a>[Cosseno](#toc0_)

&emsp;&emsp; Vimos que

$$
\cos{(a + b)} = \cos{a}\cdot\cos{b} - \text{sen}\,a\cdot\text{sen}\,b
$$

&emsp;&emsp; Fazendo $\,a = x\,$ e $\,b = x\,$ na expressão anteriro, obtemos

$$
\cos{(x + x)} = \cos{x}\cdot\cos{x} - \text{sen}\,x\cdot\text{sen}\,x
$$

$$
\therefore\quad \boxed{\,\cos{(2x)} = \cos^2{x} - \text{sen}^2\,x\,}
$$

&emsp;&emsp; Podemos obter mais uma relação substituindo $\,\text{sen}^2\,x\,$ por $\,1 - \cos^2{x}\,$ nop resultado anterior:


$$
\cos{(2x)} = \cos^2{x} - \text{sen}^2\,x
$$

$$
\cos{(2x)} = \cos^2{x} - (1 - \cos^2{x})
$$

$$
\therefore\quad \boxed{\,\cos{(2x)} = 2\cos^2{x} - 1\,}
$$





## 2.3. <a id='toc2_3_'></a>[Tangente](#toc0_)

&emsp;&emsp; Vimos que

$$
\text{tg}\,{(a + b)} = \cfrac{\ \text{tg}\,a + \text{tg}\,{b}}{\ 1 - \text{tg}\,a\cdot\text{tg}\,b}
$$

$$
\text{tg}\,{(x + x)} = \cfrac{\text{tg}\,x + \text{tg}\,x}{1 - \text{tg}\,x\cdot\text{tg}\,x}
$$

$$
\therefore\quad \boxed{\,\text{tg}\,{(2x)} = \cfrac{2\,\text{tg}\,x}{1 - \text{tg}^2\,x}\,}
$$

# 3. <a id='toc3_'></a>[Arcos Metade](#toc0_)

## 3.1. <a id='toc3_1_'></a>[Seno](#toc0_)

$$
\boxed{\,\cos{(2x)} = \cos^2{x} - \text{sen}^2\,x\,}
$$

$$
\cos{\theta} = \cos^2{\cfrac{\theta}{2}} - \text{sen}^2\,\cfrac{\theta}{2}
$$

$$
\cos{\theta} = 1 - \text{sen}^2\,\cfrac{\theta}{2} - \text{sen}^2\,\cfrac{\theta}{2}
$$

$$
\cos{\theta} = 1 - 2\,\text{sen}^2\,\cfrac{\theta}{2}
$$

$$
2\,\text{sen}^2\,\cfrac{\theta}{2} = 1 - \cos{\theta}
$$

$$
\text{sen}^2\,\cfrac{\theta}{2} = \cfrac{1 - \cos{\theta}}{2}
$$

$$
\boxed{\,\text{sen}\,\cfrac{\,\theta\,}{2} = \pm\sqrt{\cfrac{1 - \cos{\theta}}{2}}\,}
$$

## 3.2. <a id='toc3_2_'></a>[Cosseno](#toc0_)

$$
\boxed{\,\cos{(2x)} = 2\cos^2{x} - 1\,}
$$

$$
\cos{\theta} = 2\cos^2{\Big(\cfrac{\theta}{2}\Big)} - 1
$$

$$
2\cos^2{\Big(\cfrac{\theta}{2}\Big)} = 1 + \cos{\theta}
$$

$$
\cos^2{\Big(\cfrac{\theta}{2}\Big)} = \cfrac{1 + \cos{\theta}}{2}
$$

$$
\boxed{\,\cos{\Big(\cfrac{\,\theta\,}{2}\Big)} = \pm\sqrt{\cfrac{1 + \cos{\theta}}{2}}\,}
$$

## 3.3. <a id='toc3_3_'></a>[Tangente](#toc0_)

$$
\text{tg}\,{\Big(\cfrac{\theta}{2}\Big)} = \cfrac{\text{sen}\,\Big(\cfrac{\theta}{2}\Big)}{\cos{\Big(\cfrac{\theta}{2}\Big)}}
$$

$$
\text{tg}\,{\Big(\cfrac{\theta}{2}\Big)} = \cfrac{\pm\sqrt{\cfrac{1 - \cos{\theta}}{2}}}{\pm\sqrt{\cfrac{1 + \cos{\theta}}{2}}}
$$

$$
\text{tg}\,{\Big(\cfrac{\theta}{2}\Big)} = \pm\sqrt{\cfrac{\cfrac{1 - \cos{\theta}}{2}}{\cfrac{1 + \cos{\theta}}{2}}}
$$

$$
\boxed{\,\text{tg}\,{\Big(\cfrac{\,\theta\,}{2}\Big)} = \pm\sqrt{\cfrac{1 - \cos{\theta}}{1 + \cos{\theta}}}\,}
$$
