# 1. <a id='toc1_'></a>[Conceitos e Definições](#toc0_)

## 1.1. <a id='toc1_1_'></a>[Definição. <sub><sup>(Círculo Trigonométrico)</sup></sub>](#toc0_)

Consideremos sobre um plano um sistema cartesiano ortogonal com origem no ponto $O$. Um __Círculo Trigonométrico__ é definido pela circunferência $\lambda$ de raio unitário e centro na origem $O$. O comprimento dessa circnferência é $2\pi$, pois $\,r = 1\,$.

| |
| :-: |
|  |
| Fonte: Iezzi |

A cada número real $x$, com $\,0 \leqslant x < 2\pi\,$, podemos associar a um único ponto $P$ de $\lambda$ do seguinte modo:

1. Se $\,x = 0\,$, então $P$ coincide com $A$;

2. Se $\,x > 0\,$, então realizamos a partir de $A$ um percurso de comprimento $x$, no sentido anti-horário, e marcamos $P$ como ponto final do percurso;

3. Se $\,x < 0\,$, então realizamos a partir de $A$ um percurso de comprimento $|x|$, no
sentido horário. O ponto final do percurso é $P$.


# 2. <a id='toc2_'></a>[Seno](#toc0_)

## 2.1. <a id='toc2_1_'></a>[Definição](#toc0_)



Dado um número real $\,x \in [0, 2\pi]\,$, seja $P$ sua imagem no ciclo trigonométrico.

| |
| :-: |
|   |
| Fonte: Iezzi |

Denominamos __seno__ de $x$ (e indicamos $\text{sen }x$) a ordenada $\,OP_1\,$ do ponto $P$, onde $\,OP_1\,$ é a projeção ortogonal do ponto $P$ sobre o eixo das ordenadas.

## 2.2. <a id='toc2_2_'></a>[Propriedades](#toc0_)

1. Se $x$ é do primeiro ou do segundo quadrante, então $\,\text{sen } x\,$ é positivo.

2. Se $x$ é do terceiro ou do quarto quadrante, então $\,\text{sen } x\,$ é negativo.

3. Se $x$ percorre o primeiro ou o quarto quadrante, então $\,\text{sen } x\,$ é crescente.

4. Se $x$ percorre o segundo ou o terceiro quadrante, então $\,\text{sen } x\,$ é decrescente.

# 3. <a id='toc3_'></a>[Cosseno](#toc0_)

## 3.1. <a id='toc3_1_'></a>[Definição](#toc0_)

Dado um número real $\,x \in [0, 2\pi]\,$, seja $P$ sua imagem no ciclo trigonométrico.

| |
| :-: |
|  |
| Fonte: Iezzi |

Denominamos __cosseno__ de $x$ (e indicamos $\cos x$) a abscissa $\,OP_2\,$ do ponto $P$, onde $\,OP_2\,$ é a projeção ortogonal do ponto $P$ sobre o eixo das abscissas.

## 3.2. <a id='toc3_2_'></a>[Propriedades](#toc0_)

1. Se $x$ é do primeiro ou do quarto quadrante, então $\,\cos x\,$ é positivo.

2. Se $x$ é do segundo ou do terceiro quadrante, então $\,\cos x\,$ é negativo.

3. Se $x$ percorre o primeiro ou o segundo quadrante, então $\,\cos x\,$ é decrescente.

4. Se $x$ percorre o terceiro ou o quarto quadrante, então $\,\cos x\,$ é crescente.

# 4. <a id='toc4_'></a>[Tangente](#toc0_)

## 4.1. <a id='toc4_1_'></a>[Definição](#toc0_)

Dado um número real $\,x \in [0, 2\pi]\,$, $\,x \neq \cfrac{\pi}{2}\,$ e $\,x \neq \cfrac{3\pi}{2}\,$, seja $P$ sua imagem no ciclo trigonométrico. Consideremos a reta $\,\overleftrightarrow{OP}\,$ e seja T sua interseção com o eixo das tangentes.

Denominamos tangente de $x$ (e indicamos $\text{tg }x$) a medida algébrica do segmento $\,\overline{AT}\,$, copnforme a figura a seguir:

| :-: |
|  |
| Fonte: Iezzi |

Notemos que, para $\,x = \cfrac{\pi}{2}\,$, $P$ está em $B$ e, para $\,x = \cfrac{3\pi}{2}\,$, $P$ está em $B'$, então a reta $\,\overleftrightarrow{OP}\,$ fica paralela ao eixo das tangentes. Como neste caso não existe o ponto $T$, a $\,\text{tg }x\,$ não está definida.

## 4.2. <a id='toc4_2_'></a>[Propriedades](#toc0_)

1. Se $x$ é do primeiro ou do terceiro quadrante, então $\,\text{tg }x\,$ é positiva.

2. Se $x$ é do segundo ou do quarto quadrante, então $\,\text{tg }x\,$ é negativa.

3. Se $x$ percorre qualquer um dos quatro quadrantes, então $\,\text{tg }x\,$ é crescente.

## 4.3. <a id='toc4_3_'></a>[Teorema. ($\text{tg }x = \frac{\text{sen }x}{\cos x}$)](#toc0_)

Para todo $x$ real, $\,x \in [0, 2\pi]\,$, e $\,x \notin \left\{\cfrac{\pi}{2}, \cfrac{3\pi}{2}\right\}\,$, vale a relação:

$$
\boxed{\ \text{tg }x = \cfrac{\text{sen }x}{\cos x}\ }
$$

__*Demonstração.*__

| |
| :-: |
|  |
| Fonte: Iezzi |

a) Se $\,x \notin \{0, \pi, 2\pi\}\,$, a imagem de $x$ é distinta de $A$, $A'$, $B$ e $B'$. Então, temos:

$$
\begin{align*}
\triangle OAT &\sim \triangle OP_2P \\
\cfrac{|AT|}{|OA|} &= \cfrac{P_2P}{OP_2} \\
\end{align*}
$$

$$
\boxed{\ \text{tg }x = \cfrac{\text{sen }x}{\cos x}\ }
$$

b) Se $\,x \in \{0, \pi, 2\pi\}\,$, temos:

$$
\boxed{\ \text{tg }x = 0 = \cfrac{\text{sen }x}{\cos x}\ }
$$
