(ACAFE 2021) &nbsp; O número de soluções da equação $\,2\cos^2{x} - \text{sen}\,{x} = 1\,$ no intervalo $\,[0,2\pi]\,$ é

a) 2<br>
b) 1<br>
c) 3<br>
d) nenhum

&emsp; Pelo Teorema Fundamental da Trigonometria,
$$
\text{sen}^2\,{x} + \cos^2{x} = 1 \implies \cos^2{x} = \boxed{\,1 - \text{sen}^2\,{x}\,}
$$

$$
2(1 - \text{sen}^2\,{x}) - \text{sen}\,{x} - 1 = 0
$$

$$
2 - 2\,\text{sen}^2\,{x} - \text{sen}\,{x} - 1 = 0
$$

$$
-2\,\text{sen}^2\,{x} - \text{sen}\,{x} + 1 = 0 \\[12pt]
2\,\text{sen}^2{x} + \text{sen}\,{x} - 1 = 0
$$

Fazendo $m = \text{sen}\,{x}$,

$$
2\,m^2 + m - 1 = 0
$$

Pela Fórmula de "Bhaskara",

$$
m = \cfrac{-1\pm\sqrt{1^2-4\cdot 2\cdot (-1)}}{2\cdot 2} \\
$$

$$
m = \cfrac{-1\pm\sqrt{9}}{4} = \cfrac{-1\pm 3}{4} \\[10pt]
$$

$$
m = \cfrac{2}{4} = \cfrac{1}{2} \quad\text{ou}\quad m = \cfrac{-4\,}{4} = -1
$$

Devemos lembrar que:
$$
m = \text{sen}\,x
$$

$$
\text{sen}\,x = \cfrac{1}{\,2} \quad\text{ou}\quad \text{sen}\,x = -1
$$

$$
X_1 = \text{arcsen}\,\cfrac{1}{\,2} = \left\{\cfrac{\pi}{6}\,,\cfrac{5\pi}{6}\right\} \\[10pt]

X_2 = \text{arcsen}\,-1 = \left\{\cfrac{3\pi}{2}\right\}
$$
