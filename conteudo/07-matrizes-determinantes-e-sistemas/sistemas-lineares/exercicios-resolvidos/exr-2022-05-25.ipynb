EXERCÍCIO RESOLVIDO 2022-05-25

$$
\left \{
  \begin{array}{ccccccc}
  x & + & y & & & = & 2 \\
  2x & - & y & + & 3z & = & 4 \\
  -x & & & + & z & = & 0
  \end{array}
\right.
$$

__*Resolução*__.

Matriz aumentada:

$$
\left [
  \begin{array}{rrr:r}
  1 & 1 & 0 & 2 \\
  2 & -1 & \ 3 & 4 \\
  -1 & 0 & \ 1 & 0 
  \end{array} 
\right ]
$$
&emsp; $L_2 \leftarrow L_2 - 2L_1:$
$$
\left [
  \begin{array}{rrr:r}
  1 & 1 & 0 & 2 \\
  0 & -3 & \ 3 & 0 \\
  -1 & 0 & \ 1 & 0 
  \end{array} 
\right ]
$$

&emsp; $L_3 \leftarrow L_3 + L_1:$

$$
\left [
  \begin{array}{rrr:r}
  1 & 1 & 0 & 2 \\
  0 & -3 & 3 & 0 \\
  0 & 1 & 1 & 2 
  \end{array} 
\right ]
$$

&emsp; $L_3 \leftarrow 3L_3 + L_2:$

$$
\left [
  \begin{array}{rrr:r}
  1 & 1 & 0 & 2 \\
  0 & -3 & 3 & 0 \\
  0 & 0 & 6 & 6 
  \end{array} 
\right ]
$$

&emsp; $L_2 \leftarrow \cfrac{L_2}{3}\ $ e $\ L_3 \leftarrow \cfrac{L_3}{6} :$

$$
\left [
  \begin{array}{rrr:r}
  1 & 1 & 0 & 2 \\
  0 & -1 & 1 & 0 \\
  0 & 0 & 1 & 1 
  \end{array} 
\right ]
$$

$$
\text{Resp.:}
\begin{bmatrix}
x \\ y \\ z
\end{bmatrix}
=
\begin{bmatrix}
1 \\ 1 \\ 1
\end{bmatrix}
$$
