__ENEM-2024-v166__. &nbsp; As receitas anuais obtidas por uma indústria no período de 2014 a 2021, em milhão de reais, foram registradas, por pontos, em um gráfico. Nele, também está representada a reta que descreve a tendência de evolução das receitas. Essa reta pode ser utilizada para estimar as receitas dos anos seguintes.



&emsp; A estimativa da receita, em milhão de reais, dessa indústria, para o ano de 2026, obtida a partir dessa reta de tendência, é

a) 7.<br>
b) 8. &check;<br>
c) 9.<br>
d) 10.<br>
e) 11.<br>

__RESOLUÇÃO__:

&emsp; Através do gráfico apresentado, podemos tomar os pontos $\,A(2014, 4)\,$ e $\,B(2020, 6)\,$ para determinar a função $f(x) = ax + b$ que descreve a reta de tendência.

&emsp; Assim sendo,

$$
\begin{align*}
a = \cfrac{\Delta y}{\Delta x} = \cfrac{y_B - y_A}{x_B - x_A} = \cfrac{6 - 4}{2020 - 2014} = \cfrac{2}{6} = \cfrac{1}{3}\,\cdot \\[16pt]
\end{align*}
$$

$$
\begin{align*}
y = ax + b \implies b = y - ax \implies \\[10pt]
b = 4 - \cfrac{1}{3} \cdot 2014 = \cfrac{12 - 2014}{3} = -\cfrac{2002}{3}\,\cdot \\[16pt]
\end{align*}
$$

$$
\therefore\quad f(x) = \cfrac{x}{3} - \cfrac{2002}{3} \implies f(2026) = \cfrac{2026 - 2002}{3} = \cfrac{24}{3} = 8 \text{ milhões de reais}.
$$

&emsp; Resposta: b)
