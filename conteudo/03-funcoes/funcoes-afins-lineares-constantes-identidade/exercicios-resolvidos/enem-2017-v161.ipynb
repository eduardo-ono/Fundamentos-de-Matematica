__ENEM-2017-v160__. &nbsp; A base de cálculo do imposto de renda é a parte dos rendimentos recebidos pelo contribuinte sobre a qual incide o imposto. Ela é obtida após serem descontadas, dos rendimentos, as deduções legais.

&emsp; No ano de 2008, se a base de cálculo de um contribuinte teve um valor de até $\text{R\$ } 16\,473{,}72$, 0 contribuinte foi isento do imposto de renda. Se a base de cálculo ficou entre $\text{R\$ } 16\,473{,}72$ e $\text{R\$ } 32\,919{,}00$, o imposto devido foi de $15\%$ sobre o que excedeu $\text{R\$ } 16\,473{,}72$. Por fim, se a base de cálculo ultrapassou $\text{R\$ } 32\,919,00$, 0 imposto devido é dado pela soma de $\text{R\$ } 2\,466{,}79$ (correspondendo a $15\%$ da diferença $32\,919{,}00 - 16\,473{,}72$) mais $27,5\%$ do que excedeu $\text{R\$ } 32\,919{,}00$.

&emsp; O gerente de um escritório de contabilidade pediu a um estagiário que identificasse o gráfico que descrevia o valor do imposto devido, para o ano de 2008, como função da base de cálculo, apresentando-lhe cinco gráficos, sem qualquer outra informação ou valores numéricos.
