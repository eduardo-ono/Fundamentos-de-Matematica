__ENEM-2023-v139__. &nbsp; A figura representa uma escada com trs degraus, construda em concreto macio, com suas medidas especificadas.

<h4 align="center">



</h4>

&emsp; Nessa escada, pisos e espelhos tm formato retangular, e as paredes laterais tm formato de um polgono cujos lados adjacentes so perpendiculares. Pisos, espelhos e paredes laterais sero revestidos em cermica.

&emsp; A rea a ser revestida em cermica, em metro quadrado, mede

A) 1,20.<br>
B) 1,35.<br>
C) 1,65.<br>
D) 1,80.<br>
E) 1,95. &nbsp;&check;<br>

&nbsp;

__RESOLUO__

&emsp; A rea $A$ a ser revestida  igual a:

$$
A = 3 \cdot A_\text{piso} + 3 \cdot A_\text{espelho} + 2 \cdot A_\text{parede lateral}
$$

&emsp; onde

$$
\begin{align*}
A_\text{piso} & = 1,0 \cdot 0,25 = 0,25 \text{ m$^2$} \\
A_\text{espelho} & = 1,0 \cdot 0,2 = 0,2 \text{ m$^2$} \\
A_\text{parede lateral} & = 6 \cdot 0,2 \cdot 0,25 = 0,3 \text{ m$^2$}
\end{align*}
$$

&emsp; Portanto,

$$
A = 3 \cdot 0,25 + 3 \cdot 0,2 + 2 \cdot 0,3 = 0,75 + 0,6 + 0,6 = 1,95 \text{ m$^2$}
$$
