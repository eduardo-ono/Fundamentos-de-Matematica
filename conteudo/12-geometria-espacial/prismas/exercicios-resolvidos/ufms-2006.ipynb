__UFMS-2006__. &nbsp; Durante uma forte chuva, uma calha, em forma de prisma reto, de $10\,\text{m}$ de comprimento e seco transversal trapezoidal issceles de base maior $80\,\text{cm}$, base menor $60\,\text{cm}$ e profundidade $80\,\text{cm}$, como na figura abaixo, enche de gua. Se $V(x)$  a funo que define o volume de gua na calha, em cm<sup>3</sup>, em relao  profundidade $x$, em centmetros, determine $V(x)$.

__RESOLUO__

$$

V(x) & = A_{base} \cdot h \\[12pt]

V(x) & = \left[ \cfrac{(B + b) \cdot x}{2} \right] \cdot 1000 = \\[10pt]

& = \left( \frac{x + 480}{4} \right) \cdot 500x \\[10pt]

\therefore\quad V(x) & = 125x^2 + 60\,000x
\end{align*}
