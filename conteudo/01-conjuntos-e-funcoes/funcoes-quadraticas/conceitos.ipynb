Definição. Uma aplicação $f:\mathbb{R} \rightarrow \mathbb{R}$ recebe o nome de __função quadrática__ ou do __2º grau__ quando associa a cada $x \in \mathbb{R}$ o elemento $(ax^2 + bx + c) \in \mathbb{R}$ tal que $a$, $b$ e $c$ são números reais dados e $a \neq 0$.

### Exemplo

Seja a função $f(x) = x^2 - 5 \cdot x + 6$. Então, $a = 1$, $b = -5$ e $c = 6$.

Os zeros da função $f$ são dados por:

$$x = \frac{-(-5) \pm \sqrt{b^2 - 4 \cdot a \cdot c}}{2 \cdot a}$$


```python
import numpy as np
import matplotlib.pyplot as plt

a = 1
b = -5
c = 6
x_ = []
y_ = []

for x in np.arange(0, 5, 0.1):
    y = a*x**2 + b*x + c
    x_.append(x)
    y_.append(y)
    
fig = plt.figure()
axes = fig.add_subplot(111)
# Especifica o intervalo do eixo y
plt.ylim([-1, 6.5])
# Desenha o eixo x (y = 0)
plt.axhline(y=0.0, color='purple', linestyle='-', linewidth=1)
axes.plot(x_, y_)
plt.show()
```
