# 2. <a id='toc2_'></a>[Conceitos e Definições](#toc0_)

## 2.1. <a id='toc2_1_'></a>[Definição. <sub><sup>(Feixe de Retas Paralelas)</sup></sub>](#toc0_)

Um __feixe de retas paralelas__ é um conjunto de retas coplanares paralelas entre si.

## 2.2. <a id='toc2_2_'></a>[Definição. <sub><sup>(Trasversal de feixe de retas paralelas)</sup></sub>](#toc0_)

Uma __transversal de feixe de retas paralelas__ é uma reta do plano do feixe que concorre com todas as retas do feixe.

# 3. <a id='toc3_'></a>[Teorema de Tales](#toc0_)

Se duas retas são transversais de um feixe de retas paralelas, então a razão entre dois segmentos __quaisquer__ de uma das transversais é igual à razão entre os respectivos segmentos correspondentes da outra transversal.

| |
| :-: |
|  |


__*Demonstração.*__

Hipótese: $r /\negthickspace/ s /\negthickspace/ t$, $a$ e $b$ tranversais.

Tese: $\,\cfrac{AB}{BC} = \cfrac{A'B'}{B'C'}\,$ (segmentos adjacentes) e $\,\cfrac{AB}{CD} = \cfrac{A'B'}{C'D'}\,$ (segmentos não adjacentes).

Considerando o triângulo $AB'C$, a ceviana (qualquer segmento que parte de um vértice de um triângulo e corta o lado oposto a esse vértice) $\overline{BB'}$ determina os triângulos $AB'B$, de área $A_1$, e o triângulo $BB'C$, de área $A_2$. Temos, então:

$$
\cfrac{A_1}{A_2} = \cfrac{\cfrac{\overline{AB} \cdot h}{2}}{\cfrac{\overline{BC} \cdot h}{2}} = \cfrac{\ \overline{AB}\ }{\overline{BC}}\ .
$$

Analogamente, para os triângulos $A'BB'$ e $B'BC'$, temos:

$$
\cfrac{A_1}{A_2} = \cfrac{\cfrac{\overline{A'B'} \cdot h}{2}}{\cfrac{\overline{B'C'} \cdot h}{2}} = \cfrac{\ \overline{A'B'}\ }{\overline{B'C'}}\ .
$$

Portanto,

$$
\boxed{\ \cfrac{\ \overline{AB}\ }{\ \overline{BC}\ } = \cfrac{\ \overline{A'B'} }{\ \overline{B'C'}\ }\ }
$$

Alguns autores demonstram esse Teorema para o caso de dois segmentos não contínuos em cada transversal.

Analogamente ao que foi demonstrado, podemos obter:

$$
\begin{align*}
\cfrac{\ \overline{BC}\ }{\ \overline{CD}\ } &= \cfrac{\ \overline{B'C'}\ }{\ \overline{C'D'}\ } \\
\ \\
\cfrac{\ \overline{AB}\ }{\ \overline{BC}\ } \cdot \cfrac{\ \overline{BC}\ }{\ \overline{CD}\ } &= \cfrac{\ \overline{A'B'}\ }{\ \overline{B'C'}\ } \cdot \cfrac{\ \overline{B'C'}\ }{\ \overline{C'D'}\ } \\
\end{align*}
$$

$$
\boxed{\ \cfrac{\ \overline{AB}\ }{\ \overline{CD}\ } = \cfrac{\ \overline{A'B'}\ }{\ \overline{C'D'}\ }\ }
$$


# 4. <a id='toc4_'></a>[Aplicações](#toc0_)

## 4.1. <a id='toc4_1_'></a>[Conversão de escalas lineares](#toc0_)



Na escala Celsius, o ponto de fusão do gelo e da ebulição da água é 0°C e 100°C, respectivamente. Na escala Farenheigt, as respectivas temperaturas são 32°F  e 212°F.
Sabendo-se que as duas escalas são lineares, o Teorema de Tales pode ser aplicado para a coversão das respectivas temperaturas.

||
|:-:|
|  |



$$
\begin{align*}
\cfrac{C - 0}{100 - 0} &= \cfrac{F - 32}{212 - 32} \\ \\
\cfrac{C}{100} &= \cfrac{F - 32}{180} \\ \\
\cfrac{C}{5} &= \cfrac{F - 32}{9} \\
\end{align*}
$$

A partir dessa última igualdade, podemos obter $C = C(F)$ e $F = F(C)$:

$$
\boxed{C = \cfrac{5}{9}\,(F -32)}
$$

$$
\boxed{F = \cfrac{9}{5}\,C + 32}
$$
