Exercício Resolvido 107

&emsp; &emsp; Sejam $\triangle ABC$ um triângulo de lados $\,\overline{AB} = 3\,$ e $\,\overline{AC} = 5\,$ e o segmento $\overline{AF} = 3$, com $\,F \in \overline{BC}\,$. O lado $\overline{BC}$ também um dos lados de um retângulo $BCDE$, sendo que o outro lado do retângulo possui comprimento igual ao do segmento $\overline{FC}$. Determine a área do retângulo $BCDE$.

__*Resolução*__:



&emsp; &emsp; Considerações:

* O triângulo $\triangle ABC$ não é único, ou seja, não é possível determinar a medida numérica dos segmentos $\overline{BC}$ e $\overline{FC}$. 
* A área deve ser determinada pelo produto $\,\overline{BC} \cdot \overline{CD}\,$.

&emsp; &emsp; Uma forma de correlacionar as medidas dos segmentos $\overline{AB}$ e $\overline{AC}$ é através da altura $\,h = \overline{AM}\,$ relativa ao lado $\overline{BC}$, formando os triângulos $\triangle ABM$, $\triangle AMF$ e $\triangle AMC$, todos retângulos.

&emsp; &emsp; Do $\triangle AMF$ temos:

$$
\begin{align*}
\overline{AF}^2 &= h^2 + \overline{MF}^2 \quad\text{(Pitágoras)} \\
h^2 &= \overline{AF}^2 - \overline{MF}^2
\end{align*}
$$

&emsp; &emsp; Do $\triangle AMC$ temos: 

$$
\begin{align*}
\overline{AC}^2 &= h^2 + \overline{MC}^2 \quad\text{(Pitágoras)} \\
\overline{AC}^2 &= h^2 + (\overline{MF} + \overline{FC})^2 \\
\overline{AC}^2 &= \overline{AF}^2 - \overline{MF}^2 + (\overline{MF} + \overline{FC})^2 \\
(\overline{MF} + \overline{FC})^2 - \overline{MF}^2 &= \overline{AC}^2 - \overline{AF}^2 \\
(\overline{MF} + \overline{FC} + \overline{MF})(\overline{MF} + \overline{FC} - \overline{MF}) &= \overline{AC}^2 - \overline{AF}^2 \\
(2\overline{MF} + \overline{FC}) \cdot \overline{FC} &= \overline{AC}^2 - \overline{AF}^2 \\
(\overline{BF} + \overline{FC}) \cdot \overline{FC} &= 5^2 - 3^2 \\
\overline{BC} \cdot \overline{FC} &= 25 - 9 \\
\overline{BC} \cdot \overline{FC} &= 16 \\
\end{align*}
$$
