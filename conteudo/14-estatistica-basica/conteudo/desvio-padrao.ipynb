## Variância

A variância é uma medida estatística que quantifica o quanto os dados de um conjunto se dispersam em relação à média aritmética. O resultado da variância é útil para entender a variabilidade de um conjunto de dados e é a base para calcular o desvio padrão.

### Variância Populacional

Seja $x$ uma variável quantitativa e $x_1, x_2, \cdots, x_n$ os valores assumidos por $x$. Define-se a variância de $x$, denotada por $\sigma^2$, da seguinte forma: 

$$
\sigma^2 = \sum_{i = 1}^{N}\cfrac{{(x_i - \mu)^2}}{N}
$$

Exemplo:

O peso em kg de 7 adultos pertencentes a um grupo de pessoas são 55, 69, 75, 49, 92, 62, 88. Considerando que a variável quantitativa $p$ representa o peso dessas pessoas, a variância (populacional) é calculada da seguinte forma:

$$
\text{Média populacional:}\quad \mu\ =\ \cfrac{1}{N}\sum_{i=1}^N{x_i}
$$

$$
\mu\ =\ \cfrac{\,1\,}{7}\sum_{i=1}^7{p_i}\ =\ \cfrac{55 + 69 + 75 + 49 + 92 + 62 + 88}{7}\ =\ \cfrac{490}{7}\ =\ 70 \text{ kg}
$$

$$
\text{Variância populacional:}\quad \sigma^2 = \sum_{i = 1}^{N}\cfrac{{(x_i - \mu)^2}}{N}
$$

$$
\sigma^2 = \sum_{i = 1}^{7}\cfrac{(p_i - 70)^2}{7}
$$

$$
\sigma^2 = \cfrac{(55 - 70)^2 + (69 - 70)^2 + (75 - 70)^2 + (49 - 70)^2 + (92 - 70)^2 + (62 - 70)^2 + (88 - 70)^2}{7}
$$

$$
\sigma^2 = \cfrac{(-15)^2 + (-1)^2 + (5)^2 + (-21)^2 + 22^2 + (-8)^2 + 18^2}{7}
$$

$$
\sigma^2 = \cfrac{225 + 1 + 25 + 441 + 484 + 64 + 324}{7} = \cfrac{1564}{7}
$$

$$
\therefore\quad \sigma^2 = 223,43
$$

### Variância Amostral

Seja $x$ uma variável quantitativa e $x_1, x_2, \cdots, x_n$ os valores assumidos por $x$. Define-se a variância amostral de $x$, denotada por $s^2$, da seguinte forma:

$$
s^2 = \sum_{i = 1}^{N}\cfrac{{(x_i - \overline{x})^2}}{N - 1}
$$

Exemplo: Amostra com 5 elementos da população do exemplo anterior.

$$
s^2 = \sum_{i = 1}^{5}\cfrac{(p_i - 68)^2}{4}
$$

$$
s^2 = \cfrac{(55 - 68)^2 + (69 - 68)^2 + (75 - 68)^2 + (49 - 68)^2 + (92 - 68)^2}{4}
$$

$$
s^2 = \cfrac{(-13)^2 + (-1)^2 + (7)^2 + (-19)^2 + 24^2}{4}
$$

$$
s^2 = \cfrac{169 + 1 + 49 + 361 + 576}{4} = \cfrac{1156}{4}
$$

$$
\therefore\quad s^2 = 289
$$

## Desvio Padrão

O desvio padrão é uma medida estatística que quantifica a dispersão de um conjunto de dados em relação à sua média. Um desvio padrão indica que os dados estão agrupados perto da média, enquanto um valor alto significa que os dados estão mais espalhados. Ele pode ser calculado tanto para uma população inteira quanto para uma amostra representativa dela.

### Desvio Padrão Populacional

$$
\sigma = \sqrt{\sum_{i = 1}^{N}\cfrac{{(x_i - \mu)^2}}{N}}
$$

onde:

$\sigma$ = desvio padrão populacional<br>
$x_i$ = cada valor da população<br>
$\mu$ = média populacional<br>
$N$	= o tamanho da população

&emsp;&emsp; Exemplo:

&emsp; No exemplo anterior, a variância populacional dos pesos encontrada foi $\sigma^2 = 223,43$. Assim, o desvio padrão populacional é dado por

$$
\sigma = \sqrt{\sigma^2} = \sqrt{223,43}
$$

$$
\therefore\quad \sigma = 14,95
$$

### Desvio Padrão Amostral

$$
s = \sqrt{\sum_{i = 1}^{N}\cfrac{{(x_i - \overline{x})^2}}{N - 1}}
$$

&emsp; Exemplo:

&emsp; No exemplo anterior, a variância amostral dos pesos encontrada foi $s^2 = 289$. Assim, o desvio padrão amostral é dado por

$$
s = \sqrt{s^2} = \sqrt{289}
$$

$$
\therefore\quad s = 17
$$
